**1A** <br>
Graph neural network is a kind of neural network made to work for graph data.<br>
A graph contains nodes and edges, where a node is basically an entity and edges represent relationship between different entities. GNNs are well suited for capturing the relationships and interdependencies in graph and works effectively for problems where the data is naturally represented as Graphs such as Social Network, Gene Structures, Road Networks etc.<br>
For example on social media node could represent users with features such as Age, Location etc and edges could represent friendship, for classification  output could be  interest labels for each user.<br>

Difference between FCNN (Fully connected neural network), CNN (convoluted neural networks) and GNN:<br>
1.
FCNN, Considers each input independently, and doesn't focus on relationship between the inputs.
GNNs, considers the relationship between nodes/entities of a graph and  works best on graph like data.
CNNs, works well with grid like data and captures relationship between input using convolutional filters but are not suited for graph data.<br>
2.
GNNs and CNNs use parameter sharing to reduce number of learnable parameters, whereas FCNNs dont share parameters across the input and hence lead to a large number of parameters.<br>

3.
GNNS are suited of graph like data such as Social Network data, CNNs are well suited for images and FCNN can be used to do general tasks such as classification or regression but they are not accurate.<br>

Below are key components and their function in a Graph Neural Networks<br>

Node related Features:Nodes or vertices in a graph are associated to features providing information about the entity. Example in social media network node features would be matrix of all people with features such as Age, Weight etc.<br>

Edge Features: Edge features are matrix representing relationship between the nodes. For example edges between two people on social media could represent friends, distant friends etc.<br>

Node Aggregation: Each node gathers information from its neighboring nodes to capture the local structure of graph graph. This is typically done through message passing amongst nodes.<br>

Graph Convolutional Layer: This layer combines node features with information from neighbour nodes. In social media context it could be used to weigh the importance of different friends based on the edge.<br>

Activation Function: After Aggregation step using convolutional layer an activation function is used to introduce non linearity. This helps the model capture complex relationships in the data.<br>

Pooling Layer: Pooling layer can be used to downsample the data and reduce the number of paraeters for computational efficiency.<br>

Pooling Operation: This operation aggregates information from all the nodes in the graph to show global representation. In social media context it is used to predict whether two individuals would like the same movie.<br>

Output Layer: This layer takes the global representation and produces the final prediction. In the discussed example it would output a binary prediction showing whether two individuals connected by an edge would like the movie.<br>


**1C**

Graph Neural Network process the graph data through message passing mechnism. Here nodes exchange information with teh neighbouring nodes iteratively to capture and propagate information across the graph. Nodes share information with their immidiate neighbors and the neighbour share informating to their neighbours and this process continues over multiple layers allowing the GNN to capture global patterns in the graph. Information about a node's neighbors, and their neighbors, gradually influences the node's representation.<br>

The message passing in GCN involves following steps:<br>

1. Each note in a graph is a feature vector and combinining them can represent raw data or the feature matrix.
2. FOr each Node, feature vector of neighbouring nodes are aggregated. This could be done using sum, mean, or some other function.
3. The aggregated vector is then tranformed using a neural network layer, for example linear transformation followed by non linear activation function.
4. Each node's feature vector is updated with the transformed, aggregated vector, possibly combining it with the node's original vector.
5. This process is repeated with multiple layers where each layer can gather information from a farther neighbour including their neighbour and neighbour of neighbours.

The message passing in MPNN involves following steps:<br>
MPNN processes input data through message passing, updating node embeddings at each layer using non-linear activation functions and learned weight matrices. The operation involves aggregating information from neighboring nodes and updating the node representations iteratively through the layers.

<br>
Self mechanism in transformer network is a method to weigh significance of different parts of input data. It allows the model to weigh the importance of each part of input data differently and is quite effective in understanding the context and relationships in sequences such as sequences in text.
<br>
It works by generating three vectors Query (Q), Key (K), and Value (V) for each input element such as a word.
It then computes a score by taking the dot product of its Query vector with the Key vector of every other element. This score determines how much attention to pay to other parts of the input.
The scores are then scaled down and then normalized using softmax.
Then element‚Äôs Value vector is multiplied by its softmax score and then summed up. This results in a a weighted representation of the entire input.

<br>
Multi head attention:
Multi head attention is an extension of self attention mechanism where attention process is performed multiple times in parallel. This is done  so that different heads could learn to focus on different parts of the input to capture different relationship bwteen the elements.
<br>
Given input X, Query (Q), Key (K), and Value (V) computation can be performed as follows:

1. FOr each head linearly project input X on Q, K, and V matrices specific to the head
implying Qh = XW(Q), Kh = XW(K) and Vh = XW(V) where W(Q), W(K) and W(V) are weight matrices of the h head.
2.  Apply the self-attention mechanism to each head's Q, K, and V matrices.
3. Add the output from all heads
4. Multiply the result  by an additional weight matrix to produce the final output
This approach allows the network to collectively attend to all information from different subspaces at different positions leading to richer represntation of input.

## Part 1: Implement a two-layer neural network

Write a class that constructs a two-layer neural network. The class consists of two methods, an initialization that sets up the architecture of the model, and a forward pass function given an input feature.

```python
input_size = 1 * 28 * 28  # input spatial dimension of images
hidden_size = 128         # width of hidden layer
output_size = 10          # number of output neurons

class MNISTClassifierMLP(torch.nn.Module):
    
    def __init__(self):

        super(MNISTClassifierMLP, self).__init__()

        # ------------------
        # Write your implementation here.

        self.flatten = torch.nn.Flatten(start_dim=1)


        self.fc1 = torch.nn.Linear(input_size, hidden_size)


        self.act = torch.nn.ReLU()


        self.fc2 = torch.nn.Linear(hidden_size, output_size)


        self.log_softmax = torch.nn.LogSoftmax(dim=1)

        # ------------------
    
    def forward(self, x):

        x = self.flatten(x)

        # Fully connected layer 1
        x = self.fc1(x)

        # Activation function
        x = self.act(x)

        # Fully connected layer 2
        x = self.fc2(x)


        y_output = self.log_softmax(x)

        return y_output



model_fcnn = MNISTClassifierMLP().to(DEVICE)

# sanity check
print(model_fcnn)
```

## Part 2: Implement an optimizer to train the neural net model

Write a method called `train_one_epoch` that runs one step using the optimizer.

```python
import torch.nn.functional as F
def train_one_epoch(train_loader, model, device, optimizer, log_interval, epoch):
    model.train()
    losses = []
    counter = []
    train_loss_per_epoch = 0
    num_correct = 0
    
    for i, (img, label) in enumerate(train_loader):
        img, label = img.to(device), label.to(device)
        optimizer.zero_grad()

        # Forward pass
        output = model(img)

        # Calculate the negative log likelihood loss
        loss = F.nll_loss(output, label)
        train_loss_per_epoch += loss.item()


        loss.backward()


        optimizer.step()
        pred = output.argmax(dim=1, keepdim=True)
        num_correct += pred.eq(label.view_as(pred)).sum().item()

        if (i+1) % log_interval == 0:
            losses.append(loss.item())
            counter.append(
                (i * batch_size) + img.size(0) + epoch * len(train_loader.dataset))

    return losses, counter , train_loss_per_epoch/len(train_loader), num_correct
```

## Part 3: Run the optimization procedure and test the trained model

Write a method called `test_one_epoch` that evalutes the trained model on the test dataset. Return the average test loss and the number of samples that the model predicts correctly.

```python
def test_one_epoch(test_loader, model, device):
    model.eval()
    test_loss = 0
    num_correct = 0
    
    with torch.no_grad():
        for i, (img, label) in enumerate(test_loader):
            img, label = img.to(device), label.to(device)
            optimizer.zero_grad()
            # Forward pass
            output = model(img)

            # Calculate the negative log likelihood loss
            loss = F.nll_loss(output, label)
            test_loss += loss.item()

            # Get the predicted label
            pred = output.argmax(dim=1, keepdim=True)

            # Count the number of correct predictions
            num_correct += pred.eq(label.view_as(pred)).sum().item()






            
    test_loss /= len(test_loader)
    return test_loss, num_correct
```

Explain the working of each type of layer in a CNN

1. Convolution Layer
It applies set of filters  to the input.
Each filter or Kernel convolvs across the width and height of the input calculating the dot product between the filter and input. It produces a 2D map for each filter and captures local dependencies in the input.

It has filter and stride size as hyper parameters:
Filter size determines the dimension of the kernel. Common choices are 3x3 or 5x5.
Stride size dictates the step length the filter moves across the input. A stride of 1 moves the filter one pixel at a time.

2. Activation Layer
This layer transforms the output of convolution layer using some activation function such as RELU to introduce non linearity in the network.

3. Pooling Layer
This layer reduces the spatial size of the input volume from the convolution layer, this also makes the network less sensetive to the location size of the feature.
Common pooling operations include max pooling or average pooling where either max or average of the group of pixels are taken respectively.
Pooling is applied to each slice of the input and resizes it spatially.

4. Fully Connected Layer
Fully connected layer computes the class scores resulting in the volume size equal to the number of classes in the classification problem. Each neuron in a fully connected layer is connected to every other neuron in the previous layer.
Fully connected layers are used towards the end of a CNN to classify the input to a label. This layer take the features learned by the previous layers and use them to classify an image as it takes the high dimensional output from previous convolution and pooling layers and flattens it into one dimensional vector combining all the extracted features across the image rather than focusing on the local features. This layer helps in understanding th eglobal context of the image.

5. Output Layer:
This layer transforms the input from the Fully connected layer to a probablities. for classification problem this layer can use softmax function which converts the logits into probablities where all the probablities sum up to 1 providing the class probablities.


Examples of CNN Architectures are :
LeNet-5: it is of the earliest CNNs which included two sets of convolution and average pooling layers, followed by fully connected layers. It was used to recognize hand written and machine printed characters.

AlexNet: It is a complex CNN which consists of 5 convolutional layers, some of which are followed by max-pooling layers, and 3 fully connected layers at the end. It was used for large scale image recognition task.

VGGNet: Known for its simplicity with a series of convolutional layers with small filters (3x3) followed by max-pooling layers and ending with three fully connected layers. It is also used for large scale image recognition.

ResNet: It introduced the concept of skiping connections to enable training of very deep networks. It is also used for large scale image recognition

GoogLeNet (Inception Network): Characterized by its inception modules, which use parallel convolutions . It is also used for large scale image recognition

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

input_size = 1 * 28 * 28  # input spatial dimension of images
hidden_size = 128         # width of hidden layer
output_size = 10          # number of output neurons

class CNN(torch.nn.Module):

    def __init__(self):
        super(CNN, self).__init__()
        self.flatten = torch.nn.Flatten(start_dim=1)
        # ------------------
        # Write your implementation here.
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5, stride=1, padding=0) # first convolutional layer
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5, stride=1, padding=0)
        self.relu_activation_function = nn.ReLU()
        self.pooling_layer = nn.MaxPool2d(kernel_size=2, stride=2)
        # second convolutional layer
        self.fc = nn.Linear(in_features=20 * 4 * 4, out_features=10) # third fully-connected layer
        # ------------------

    def forward(self, x):
        # Input image is of shape [batch_size, 1, 28, 28]
        # Need to flatten to [batch_size, 784] before feeding to fc1


        x = self.pooling_layer(self.relu_activation_function(self.conv1(x)))
        x = self.pooling_layer(self.relu_activation_function(self.conv2(x)))
        x = self.flatten(x)

        x = self.relu_activation_function(x)
        x = self.fc(x)
        y_output = F.log_softmax(x, dim=1)

        # ------------------
        # Write your implementation here.



        return y_output
        # ------------------

model_cnn = CNN()
model_cnn.to(DEVICE)

# sanity check
print(model_cnn)
```

```python

def train_one_epoch(train_loader, model, device, optimizer, log_interval, epoch):
    model.train()
    losses = []
    counter = []
    train_loss_per_epoch = 0
    num_correct = 0


    for i, (img, label) in enumerate(train_loader):
        img, label = img.to(device), label.to(device)

        output = model(img)
        optimizer.zero_grad()
        loss = F.nll_loss(output, label)
        train_loss_per_epoch += loss.item()
        # Backward pass
        loss.backward()

        # Update the weights
        optimizer.step()
        pred = output.argmax(dim=1, keepdim=True)
        num_correct += pred.eq(label.view_as(pred)).sum().item()




        # ------------------
        # Copy the implementation from Problem 4 here


        # ------------------

        # Record training loss every log_interval and keep counter of total training images seen
        if (i+1) % log_interval == 0:
            losses.append(loss.item())
            counter.append(
                (i * batch_size) + img.size(0) + epoch * len(train_loader.dataset))

    return losses, counter, train_loss_per_epoch/len(train_loader), num_correct
```

```python
def test_one_epoch(test_loader, model, device):
    model.eval()
    test_loss = 0
    num_correct = 0

    with torch.no_grad():
        for i, (img, label) in enumerate(test_loader):
            img, label = img.to(device), label.to(device)

            output = model(img)

            # Calculate the negative log likelihood loss
            loss = F.nll_loss(output, label)
            test_loss += loss.item()

            # Get the predicted label
            pred = output.argmax(dim=1, keepdim=True)

            # Count the number of correct predictions
            num_correct += pred.eq(label.view_as(pred)).sum().item()

            # ------------------
            # Copy the implementation from Problem 4 here

            # ------------------

    test_loss /= len(test_loader)
    return test_loss, num_correct
```

```python
from tqdm import tqdm
# Hyperparameters
lr = 0.01
max_epochs=15
gamma = 0.95

# Recording data
log_interval = 100

# Instantiate optimizer (model was created in previous cell)
optimizer = torch.optim.SGD(model_cnn.parameters(), lr=lr)

train_losses = []
train_accuracy = []
test_accuracy = []
train_counter = []
test_losses = []
train_loss_per_epoch = []
test_correct = []
train_correct = []

pbar = tqdm(range(max_epochs), desc="Epoch")
for epoch in range(max_epochs):
    train_loss, counter, train_loss_per_one_epoch, num_correct_train = train_one_epoch(train_loader, model_cnn, DEVICE, optimizer, log_interval, epoch)
    test_loss, num_correct = test_one_epoch(test_loader, model_cnn, DEVICE)

    # Record results
    train_losses.extend(train_loss)
    train_counter.extend(counter)
    test_losses.append(test_loss)
    test_correct.append(num_correct)
    train_correct.append(num_correct_train)
    train_loss_per_epoch.append(train_loss_per_one_epoch)
    test_accuracy.append((num_correct/(len(test_loader) * 64)) * 100)
    train_accuracy.append((num_correct_train/(len(train_loader) * 64)) * 100)

    pbar.set_postfix(dict={'current_accuracy_test' : (num_correct/(len(test_loader) * 64)) * 100 , 'current_accuracy_train' : ((num_correct_train/(len(train_loader) * 64)) * 100)}, refresh=True)
    pbar.update(1)
    #print(f"Test accuracy at epoch {epoch+1}: {(num_correct/len(test_loader.dataset)) * 100}% ", end='\r', flush=True)


print(f"Test accuracy Final: {(test_correct[-1]/(len(test_loader) * 64))*100}")
```

```python
# 3. Get 20 incorrect predictions in test dataset

# Collect the images, predictions, labels for the first 20 incorrect predictions
# Initialize empty tensors and then keep appending to the tensor.
# Make sure that the first dimension of the tensors is the total number of incorrect
# predictions seen so far
# Ex) incorrect_imgs should be of shape i x C x H x W, where i is the total number of
# incorrect images so far.
incorrect_imgs = torch.Tensor().to(DEVICE)
incorrect_preds = torch.IntTensor().to(DEVICE)
incorrect_labels = torch.IntTensor().to(DEVICE)

with torch.no_grad():
    # Loop over the test set batches until incorrect_imgs.size(0) >= 20
    for images, labels in test_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        # Forward pass
        output = model_cnn(images)

        # Get predictions
        pred = output.argmax(dim=1)


        # Find indices where predictions and labels don't match (incorrect predictions)
        incorrect_indices = torch.nonzero(pred != labels).squeeze()





        # Append incorrect predictions to tensors using torch.cat
        if incorrect_indices.numel() > 0:
            if incorrect_indices.numel() == 1 :
                incorrect_indices = incorrect_indices.unsqueeze(0)

            incorrect_imgs = torch.cat((incorrect_imgs, images[incorrect_indices]), dim=0)

            incorrect_preds = torch.cat((incorrect_preds, pred[incorrect_indices]), dim=0)
            incorrect_labels = torch.cat((incorrect_labels, labels[incorrect_indices]), dim=0)


        # Break the loop if we have collected at least 20 incorrect predictions
        if incorrect_imgs.size(0) >= 20:
            break


```
